# Problem 1

Fizikte, bir nesnenin sabit ivmeyle hareket ederken son hızını 
(veya hızını) bulmak için aşağıdaki denklem kullanılabilir:
```
vf = vi + at
burada:
vf= son hız
vi= ilk hız
a= hızlanma
t= zaman
```
İlk hız, ivme ve zaman verildiğinde, son hızı döndürecek bir fonksiyon yazın.

````python
#vf= son hız, vi= ilk hız, a= hızlanma , t= zaman şeklinde tanımlayalım.

def SonHiz (vi,a,t): # def yapısını kullanarak bir fonksiyon tanımladık. 
    vf = vi + (a*t) #Bu fonksiyon ne yapsın işte onu belirttik
    print("Son hızınız: ", vf)

SonHiz(10,5,4) #fonksiyonu tanımladık. 
````

# Problem 3

1'den 100'e kadar olan sayılardan pisagor üçgeni oluşturanları ekrana yazdıran bir fonksiyon yazın.(a <= 100,b <= 100)


````python
# bir kenar a, birisi b hipotenus c olursa : c = (a**2 + b**2)**0.5 seklinde hesaplanır. 

def pisagor():
    pisagorListesi=[]
    for a in range (1,101):
        for b in range (1,101):
            
            c = (a ** 2 + b ** 2) ** 0.5

            if (c == int(c)):
                pisagorListesi.append((a,b,c))
    return pisagorListesi
for x in pisagor() :
    print(x) 
````

# Problem 4

Bir duvar boyamaya karar verdiniz. Boya kutusunun üzerindeki talimatta, 1 kutu boyanın 5 metrekarelik bir duvarı boyayabileceği yazıyor. Rastgele bir duvar yüksekliği ve genişliği verildiğinde, kaç kutu boya satın almanız gerektiğini hesaplayın.

kutu sayısı = (duvar yüksekliği * duvar genişliği) ÷ kutu başına kaplama.

örneğin Yükseklik = 2, Genişlik = 4, Kaplama = 5

kutu sayısı = (2 * 4) ÷ 5 = 1.6

Ancak bir kutu boyanın 0,6'sını satın alamayacağınız için, sonuç 2 kutuya yuvarlanmalıdır .


````python

def kutuBoya(genişlik, yükseklik):
    kaplama = 5
    kutuSayısı = (genişlik * yükseklik) / kaplama
    if (type(kutuSayısı == float)):
        kutuSayısı == int(kutuSayısı)
        kutuSayısı +=1 
    print("Almanız gerken kutu sayısı: " , kutuSayısı)

kutuBoya(4,2)
````

# Problem 5

Bir kelimedeki ünlü ve ünsüz harfleri sayan bir fonksiyon yazınız.

````python
def harfler(text):
    unluHarfler = 0
    unsuzHarfler = 0
    text = text.lower() #girilen tüm harfler aynı fontta yanı kucuk olsun ki daha az şey tanımlasak cod calışması için yetsin. 

    for i in range (len(text)): 
        if text [i] in ["a","e","i","ı","o","ö","u","ü"]:
            unluHarfler +=1
        else:
            unsuzHarfler += 1

    print("Girilen metindeki ÜNLÜ Harflerin sayısı: ", unluHarfler)
    print("Girilen metindeki ÜNSÜZ Harflerin sayısı: ", unsuzHarfler)

harfler("Lokman Kara")
````

# Problem 7

0'dan 10'a kadar olan sayıların toplamını hesaplayan özyinelemeli (Recursive Function) bir fonksiyon oluşturan bir fonksiyon yazınız.

Özyinelemeli (Recursive Function) bir işlev, kendini tekrar tekrar çağıran bir işlevdir.

````python
def toplama(x):
    if x != 0 :
        return x + toplama(x - 1 )
    else:
        return 0 
toplama(10) # Bu fonsiyon aslında #print(toplama(10)) bu demektir. Fonksiyon ile daha kısa ve hızlı olmasını sağladık.
````
