# Problem 1
Kullanıcıdan bir sayı alınız. Sayı 0'dan büyükse pozitif , 0'dan küçükse negatif sıfıra eşitse sıfır yazdırınız.

```python
#input ile sayı alalım
#if ile eşitlikleri saglarız.

sayi = int(input("Bir sayı giriniz: "))
if sayi>0 :
    print("Pozitif.")
elif sayi<0 :
    print("Negatif.")
else:
    print("Girdiğiniz sayı sıfıra eşittir.")
```

 # Problem 3
Kullanıcıdan aldığınız bir sayının "Armstrong" sayısı olup olmadığını bulmaya çalışın.
Örnek olarak, Bir sayı eğer 4 basamaklı ise ve oluşturan rakamlardan herbirinin 4. kuvvetinin toplamı( 3 basamaklı sayılar için 3.kuvveti ) o sayıya eşitse bu sayıya "Armstrong" sayısı denir.

Örnek olarak : 1634 = 1^4 + 6^4 + 3^4 + 4^4

```python
# Şimdi once sayı inputlarız. # inputa girilen sayının rakamlarını tek tek buluruz. Bunun için liste yapabiliriz.  
# Bulunan her basamak sayı kadar ussu alınıp hepsini toplarız. 
# çıkan toplam sayıya eşitse ARMSTRONG sayıdır. 

toplam = 0 # toplamı once boş bi atama yapıyoruz. 
sayi = int(input ("Bir sayı giriniz")) 
basamak = str(sayi) 

for i in basamak: # i diye yeni bir liste oluşturuyoruz. bu yeni listenin elemanlarını *basamak=str(sayi) fonksiyonu ile alıyoruz.  
    rakam = int(i)**len(basamak) #int dememizin sebebi i kümesindeki sayıları kullanmak istememiz. len ise listenin uzunluğunu verir. ** işsarettinden sonra len koyarak basamak sayısı kadar üs al demiş oluyoruz. 
    toplam += rakam 
if (sayi == toplam): #toplam sayımıza eşitse eger 
    print("Amstrong sayıdır. ")
else:
    print("Armstrong sayı değildir.")
```

# Problem 4
YetGen Core Python mezunu bir öğretmenimiz öğrencilerine 1'den 10'a kadar sayılarla çarpım tablosunu öğretirken Python'dan destek almak istiyor. Bu konuda öğretmenimize yardımcı olmak ister misin?

İpucu: İç içe 2 tane for döngüsü kullanın. 

Aynı zamanda sayıları range() fonksiyonunu kullanarak elde edin.

```python
sayi = int(input("Bir sayı giriniz: "))
for i in range(1,11,1): 
    print(f"{sayi} sayısının {i}'ile çarpımı= " , i*sayi)
```

# Problem 5
Markete giren bir YetGen'li sıra beklememek için hızlı kasaya gider. Hızlı kasanın çalışma mantığı şu şekildedir:
- Kullanıcının aldığı her ürünün fiyatını tek tek girmesi gerekir.
- Kullanıcı kasadan çıkmak isterse q tuşuna basmalıdır.
- Kasa girilen fiyatları son fiyat isimli değişkene ekler.
- Değişkeni ekrana bastırır.

İpucu: while döngüsünü sonsuz koşulla başlatın ve kullanıcı q'ya basarsa döngüyü break ile sonlandırın.

```python
#
toplam = 0
while True : 
    secim = input(" Çıkmak için 'q' diğer durumlarda sayıyı giriniz: ") # q direkt metinsel oldugu icin onu int cevirmedik. ama

    if secim == 'q' : 
        print("Uygulama sonlandırıldı.")
        break 
    else: 
        sayi = int(secim) #ama secim str olarak tanımlıydı bunu integere cevirelim ki toplama yapabilelim. 
        toplam += sayi 
    print(f"Toplam tutarınız {toplam} TL'dir.")


 
```

# Problem 6
1'den 100'e kadar olan sayılardan sadece 3'e bölünen sayıları ekrana bastırın. Bu işlemi continue ile yapmaya çalışın.

```python
for i in range(1,101):
    if i % 3 == 0:
        print(i)
    else:
        continue
```

```python
for i in range(0,101,3):
    print(i)
    continue
```
